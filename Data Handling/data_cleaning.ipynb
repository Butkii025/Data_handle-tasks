```python
# data read

import pandas as pd

df=pd.read_csv("raw_data.csv")
print(df)
```

## Handle missing values

```python
# missing values

df.isnull() # return empty values (True)
df.isnull().sum() # total no. of mussing values

```

```python
df.dropna() # row drop which have no missing data value
```

```python
df.dropna(axis=1) # cloumn drop with no null element
```

```python
df.fillna(0) # all missing value ~ 0
```

```python
# calulate average of all age

cleaned_data= df.copy()
age_mean=cleaned_data["age"].mean()

cleaned_data["age"]= cleaned_data["age"].fillna(age_mean)

print(cleaned_data)

# df["age"].fillna(age_mean)  # rplace empty with mean value
```

```python
# forward filling data

df.ffill()  
```

```python
# backward filling

df.bfill()  
```

## Handle duplication

```python
# searching for duplication

df["country"].duplicated() # return Ture if duplicated data was there

df[["country","gender"]].duplicated() # if both are correct ~ Ture
```

```python
# removing duplication (de_duplication)

df.drop_duplicates()  # delete duplicates
```

```python
# converting data type

# converting age data type (float to int)

df2=df.copy() # make a copy of dataset

df2=df2.fillna(0) # replace empty with 0
df2["age"]=df2["age"].astype("int64").copy()  # actual conversion
df2.dtypes

```
